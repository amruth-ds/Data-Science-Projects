```python
import pandas as pd
import requests

def dfdata(i):
    q=[]
    l=[]
    a=[]
    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
    resp = requests.get(i, headers=headers).text
    resp_parsable=BeautifulSoup(resp,'html.parser')
    for x in resp_parsable.find_all('div',{'class':'quoteText'}):
        q.append(list(x.children)[0].strip())
    for x in resp_parsable.find_all('a',{'class':'smallText'}):
        l.append(x.text.strip().split()[0])
    for x in resp_parsable.find_all('span',{'class':'authorOrTitle'}):
        a.append(x.text.strip())
    table=pd.DataFrame({'author':a,'quote':q,'likes':l})
    return(table)
    
    
k=dfdata("https://www.goodreads.com/quotes")
k.head()
```

```python
import pandas as pd
import requests
import re

def dfbar(i):
    name=[]
    rating=[]
    author=[]
    pattern=re.compile(r'\d\.\d{2}')
    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
    resp = requests.get(i, headers=headers).text
    resp_parsable=BeautifulSoup(resp,'html.parser')
   
    for x in resp_parsable.find_all('span',{'role':'heading'}):
        name.append(x.text.strip())

    for x in resp_parsable.find_all('a',{'class':'authorName'}):
        author.append(x.text.strip())
    for x in resp_parsable.find_all('span',{'class':'minirating'}):
        rating.append(re.findall(pattern,x.text.strip())[0])
    table=pd.DataFrame({'Book Name':name,'Author':author,'Rating':rating})
    return(table)
    
    
k=dfbar("https://www.goodreads.com/list/show/19.Best_for_Book_Clubs?page=1")
k.head()
```

```python
import pandas as pd
import requests
import re
from selenium import webdriver


def dfnaukari(i):
    jobtitle=[]
    company=[]
    Exp=[]
    Location=[]
    Salary=[]
    browser=webdriver.Firefox()
    url=i
    browser.get(url)
    html=browser.page_source
    resp_parsable=BeautifulSoup(html,'html.parser')
    for x in resp_parsable.find_all('li',{'class':'desig'}):
        jobtitle.append(x.text.strip())
    print(len(jobtitle))
    for x in resp_parsable.find_all('span',{'class':'org'}):
        company.append(x.text.strip())  
    print(len(company))
    for x in resp_parsable.find_all('span',{'class':'exp'}):
        Exp.append(x.text.strip()) 
    print(len(Exp))
    for x in resp_parsable.find_all('span',{'class':'loc'}):
        Location.append(x.text.strip()) 
    print(len(Location))
    for x in resp_parsable.find_all('span',{'class':'salary'}):
        Salary.append(x.text.strip()) 
    print(len(Salary))
    table=pd.DataFrame({'Job Title':jobtitle,'Company Name':company,'Experience Required':Exp,'Location':Location,'Salary':Salary})
    return(table)
    
    
k=dfnaukari("https://www.naukri.com/data-science-jobs-1")
k.head()
```
