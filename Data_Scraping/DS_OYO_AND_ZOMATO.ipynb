```python
import pandas as pd
import numpy as np
def oyo_data(url):
    z=url
    browser.get(z)
    html=browser.page_source
    soup=BeautifulSoup(html,'html.parser')
    listings=soup.find_all("div",{'class':'hotelCardListing__descriptionWrapper'})
    n=[]
    for i in listings:
        try:
            name=i.find_all('h3')[0].text
        except:
            name=np.nan
        n.append(name)
    l=[]
    for i in listings:
        try:
            name=i.find_all('span')[0].text
        except:
            name=np.nan
        l.append(name)
    r=[]
    for i in listings:
        try:
            name=i.find_all('div',{'class':'hotelRating__wrapper'})[0].text.split()[0]
        except:
            name=np.nan
        r.append(name)
    fp=[]
    for i in listings:
        try:
            name=i.find_all('span',{'class':'listingPrice__finalPrice'})[0].text.split()[0]
        except:
            name=np.nan
        fp.append(name)
    df_oyo=pd.DataFrame({'Hotel_Name':n,'Street':l,'Rating':r,'Final_Price':fp})
    return(df_oyo)
def oyo_soup_data(data):
    html=data
    soup=BeautifulSoup(html,'html.parser')
    listings=soup.find_all("div",{'class':'hotelCardListing__descriptionWrapper'})
    n=[]
    for i in listings:
        try:
            name=i.find_all('h3')[0].text
        except:
            name=np.nan
        n.append(name)
    l=[]
    for i in listings:
        try:
            name=i.find_all('span')[0].text
        except:
            name=np.nan
        l.append(name)
    r=[]
    for i in listings:
        try:
            name=i.find_all('div',{'class':'hotelRating__wrapper'})[0].text.split()[0]
        except:
            name=np.nan
        r.append(name)
    fp=[]
    for i in listings:
        try:
            name=i.find_all('span',{'class':'listingPrice__finalPrice'})[0].text.split()[0]
        except:
            name=np.nan
        fp.append(name)
    df_oyo=pd.DataFrame({'Hotel_Name':n,'Street':l,'Rating':r,'Final_Price':fp})
    return(df_oyo)
    
k=oyo_data('https://www.oyorooms.com/hotels-in-coimbatore/?page=4')  
k
    
    
```

```python
import time
D={}
url='https://www.oyorooms.com/hotels-in-coimbatore/?page=1'
browser.get(url)
btns=browser.find_elements_by_class_name("ListingPagination__pageContainer--page")
for j in range(0,len(btns)):
    btns[j].click()
    time.sleep(10)
    D[j]=oyo_soup_data(browser.page_source)
    btns=browser.find_elements_by_class_name("ListingPagination__pageContainer--page")
```

```python
master_oyo_table=pd.concat([D[i] for i in D],axis=0)
master_oyo_table['Rating']=master_oyo_table['Rating'].astype('float')
master_oyo_table['Final_Price']=master_oyo_table['Final_Price'].str.replace('₹','').astype('float')
master_oyo_table
```
