```python
browser=webdriver.Firefox()
def tablefunc(z):
    k=z
    print(k)
    browser.get(k)
    html=browser.page_source
    soup=BeautifulSoup(html,'html.parser')
    table_terror=soup.find_all('table',{'class':'wikitable'})
    rows=table_terror[0].find_all('tr')
    Type=[]
    Dead=[]
    Injured=[]
    Location=[]
    Details=[]
    Perpetrator=[]
    Part_of=[]
    Year=[]
    Month=[]
    count=0
    for i in rows[1:]:
        Type.append(i.find_all('td')[1].text.strip('\n'))
        Dead.append(i.find_all('td')[2].text.strip('\n'))
        Injured.append(i.find_all('td')[3].text.strip('\n'))
        Location.append(i.find_all('td')[4].text.strip('\n'))
        Details.append(i.find_all('td')[5].text.strip('\n'))
        Perpetrator.append(i.find_all('td')[6].text.strip('\n'))
        try:
            Part_of.append(i.find_all('td')[7].text.strip('\n'))
        except:
            Part_of.append(' ')
        Month.append(k.split('_')[-2])
        Year.append(k.split('_')[-1])
        
    df_terror=pd.DataFrame({'Type':Type,'Dead':Dead,'Injured':Injured,'Location':Location,'Details':Details,'Perpetrator':Perpetrator,'Part of':Part_of,'Month':Month,'Year':Year})
    return(df_terror)

k=tablefunc('https://en.wikipedia.org/wiki/List_of_terrorist_incidents_in_January_2015')
k.head()

```

```python
browser=webdriver.Firefox()
browser.get('https://en.wikipedia.org/wiki/List_of_terrorist_incidents')
html=browser.page_source
soup=BeautifulSoup(html,'html.parser')
soup
```

```python
data_needed=soup.find_all('div',{'class':'div-col columns column-width'})
data_main=data_needed[0].find_all('li')
data_main
```

```python
ls=[]
count=0
for x in data_main[49:]:
    for i in (x.find_all('a',{'href':True})):
        ls.append('https://en.wikipedia.org'+i['href'])
```

```python
D={}
for i in range(len(ls)):
    D[i]=tablefunc(ls[i])
```

```python
big_listterror_book_table=pd.concat([D[i] for i in D],axis=0)
big_listterror_book_table['country']=big_listterror_book_table['Location'].map(lambda x:x.split(",")[-1])
big_listterror_book_table
```

```python
import re
import numpy as np
X=[]
for k in big_listterror_book_table['Dead']:
    k = k.rstrip('+')
    x = re.findall(r'\d{1,4}', k)
    if(len(x)>1):
        X.append(x[1])
    else:
        if(len(x)==0):
            X.append(np.nan)
        else:
            X.extend(x)

X=[float(i) for i in X]
```

```python
big_listterror_book_table['d_cleaned']=X
big_listterror_book_table
```

```python
big_listterror_book_table.groupby('Month')['d_cleaned'].mean()
```

```python
big_listterror_book_table
```

```python
import re
import numpy as np
Y=[]
for k in big_listterror_book_table['Injured']:
    k = k.rstrip('+')
    x = re.findall(r'\d{1,4}', k)
    if(len(x)>1):
        Y.append(x[1])
    else:
        if(len(x)==0):
            Y.append(np.nan)
        else:
            Y.extend(x)

Y=[float(i) for i in Y]
```
