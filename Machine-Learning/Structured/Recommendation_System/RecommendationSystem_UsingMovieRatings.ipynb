# Preparing prediction model using KNNWithMeans

```python
import pandas as pd 
import numpy as np
from surprise import KNNWithMeans
from surprise import Dataset
from surprise import accuracy
from surprise import Reader
from surprise.model_selection import train_test_split
```

```python
DF_Ratings
```

```python
#Rating range and dataset is considered in this block
viewer_rating=Reader(rating_scale=(0,5))
Data_main=Dataset.load_from_df(DF_Ratings,viewer_rating)
```

```python
#Train and test dataset prepared so to compute RMSE
[trainset,testset]=train_test_split(Data_main,test_size=.20,shuffle=True)
```

```python
# Model Developed which can predict us all the possible scenarios
recom=KNNWithMeans(k=50,sim_options={'name':'cosine','user_based':True})
recom.fit(Data_main.build_full_trainset())
test_pred = recom.test(testset)
RMSE = accuracy.rmse(test_pred)
```

```python
#Using list computing the all predictions for a particular user which user didn't watch(Code testing)
list_data=[]
for i in Ratings_pivot.index:
    if(Ratings_pivot[2][i]==0 and recom.predict(2,i).est>2.5):
        list_data.append((2,i,recom.predict(2,i).est))
```

```python
list_data
```

```python
#User Defined function to sort and compute TOP 10 movies for any particular user
def Sort_TOP_10(sub_li): 
    sub_li.sort(key = lambda x: x[2],reverse=True) 
    return sub_li[:10] 

print(Sort_TOP_10(list_data))
```

```python
#Final list preparation which compute Top 10 predictions for all the users with rating above 2.5
list_data=[]
list_full_data=[]
for j in Ratings_pivot.columns:
    list_data=[]
    for i in Ratings_pivot.index:
        if(Ratings_pivot[j][i]==0 and recom.predict(j,i).est>2.5):
            list_data.append((j,i,recom.predict(j,i).est))
    list_full_data.extend(Sort_TOP_10(list_data))
```

```python
list_full_data
```

```python
#Preparing the data set for the list of top 10 predictions computed above
user_id=[]
movie_id=[]
rating=[]
for i in list_full_data:
    user_id.append(i[0])
    movie_id.append(i[1])
    rating.append(i[2])
    
UserMovieRating=pd.DataFrame({'User_ID':user_id,'Movie_ID':movie_id,'Rating':rating})
                 
```
