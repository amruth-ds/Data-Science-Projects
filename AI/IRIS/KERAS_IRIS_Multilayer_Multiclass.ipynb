```python
from keras.models import Sequential
from keras.layers import Flatten
from keras.layers import Dense
```

```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
data=pd.read_csv('C:/Users/Amruth PC/Desktop/SEM 2/AI/data/iris.csv')
```

```python
data.head()
```

```python
X_input=data[['petallength', 'petalwidth', 'sepallength', 'sepalwidth']]
Y_label = data[['species']]
```

```python
mymodel = Sequential()
```

```python
mymodel.add(Dense(units=300,input_dim=4,activation ='relu'))
mymodel.add(Dense(units=200,activation='relu'))
mymodel.add(Dense(units=50,activation='relu'))
```

```python
mymodel.add(Dense(units=3,activation='softmax'))
```

```python
mymodel.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
# fit model
mymodel.fit(X_input,Y_label,epochs=2000)
```

```python
mymodel.evaluate(X_input,Y_label)
```
